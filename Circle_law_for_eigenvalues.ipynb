# Part 5: Circle law for eigenvalues

# part A:

# i.

```python
import numpy as np 
import matplotlib.pyplot as plt
n=[20,50,100,200,500,1000]
def Gaussian_dist(n):
    for i in range(0,len(n)):
        M=(np.array(np.random.normal(0,1/np.sqrt(n[i]),size=(n[i]**2)))).reshape(n[i],n[i])
        Landa=np.linalg.eigvals(M)
        plt.figure()
        plt.scatter(Landa.imag,Landa.real,color="orange")
        plt.xlabel("Re(Landa)")
        plt.ylabel("Im(Landa)")
        plt.title(f"for n={n[i]}")
        angle=np.linspace(0,2*np.pi,1000)
        plt.plot(np.cos(angle),np.sin(angle),color="purple",linewidth=4)
Gaussian_dist(n)
```

# ii.

```python
def Uniform_dist(n):
    for i in range(0,len(n)):
        a=-np.sqrt(3/n[i])
        b=np.sqrt(3/n[i])
        M=(np.array(np.random.uniform(a,b,size=(n[i]**2)))).reshape(n[i],n[i])
        Landa=np.linalg.eigvals(M)
        plt.figure()
        plt.scatter(Landa.imag,Landa.real,color="green")
        plt.xlabel("Re(Landa)")
        plt.ylabel("Im(Landa)")
        plt.title(f"for n={n[i]}")
        angle=np.linspace(0,2*np.pi,1000)
        plt.plot(np.cos(angle),np.sin(angle),color="red",linewidth=4)
Uniform_dist(n)
```
